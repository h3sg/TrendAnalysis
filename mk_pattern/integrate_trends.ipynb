```python
import json
```

```python
def integrate_word_count(dictionary,file):
    with open(file,'r',encoding='utf-8') as f:
        data=json.load(f)
    for kw in data.keys():
        dictionary[kw]=[0]*4
        if data[kw]=='上升趋势':
            dictionary[kw][0]=1
        if data[kw]=='下降趋势':
            dictionary[kw][0]=2
        if data[kw]=='没有趋势':
            dictionary[kw][0]=3
    return dictionary
```

```python
def integrate_trends(dictionary,file1,file2,file3,file4):
    with open(file1,'r',encoding='utf-8') as f1:
        data1=json.load(f1)
    for kw1 in data1.keys():
        dictionary[kw1]=[0]*4
        if data1[kw1]=='上升趋势':
            dictionary[kw1][0]=1
        if data1[kw1]=='下降趋势':
            dictionary[kw1][0]=2
        if data1[kw1]=='没有趋势':
            dictionary[kw1][0]=3
    with open(file2,'r',encoding='utf-8') as f2:
        data2=json.load(f2)
    for kw2 in data2.keys():
        if kw2 in dictionary:
            if data2[kw2]=='上升趋势':
                dictionary[kw2][1]=1
            if data2[kw2]=='下降趋势':
                dictionary[kw2][1]=2
            if data2[kw2]=='没有趋势':
                dictionary[kw2][1]=3
    with open(file3,'r',encoding='utf-8') as f3:
        data3=json.load(f3)
    for kw3 in data3.keys():
        if kw3 in dictionary:
            if data3[kw3]=='上升趋势':
                dictionary[kw3][2]=1
            if data3[kw3]=='下降趋势':
                dictionary[kw3][2]=2
            if data3[kw3]=='没有趋势':
                dictionary[kw3][2]=3
    with open(file4,'r',encoding='utf-8') as f4:
        data4=json.load(f4)
    for kw4 in data4.keys():
        if kw4 in dictionary:
            if data4[kw4]=='上升趋势':
                dictionary[kw4][3]=1
            if data4[kw4]=='下降趋势':
                dictionary[kw4][3]=2
            if data4[kw4]=='没有趋势':
                dictionary[kw4][3]=3
    return dictionary
```

```python
from itertools import product
```

```python
def count_trends(readfile,writefile):
    with open(readfile,'r',encoding='utf-8') as f1:
        data=json.load(f1)
    all_keys = [''.join(map(str, p)) for p in product([1, 2, 3], repeat=4)]
    result = {key: [] for key in all_keys}
    for kw, values in data.items():
        key = ''.join(map(str, values))  # 将 [1, 2, 2, 3] 转换为 '1223'
        if key in result:
            result[key].append(kw)
    with open(writefile, 'w',encoding='utf-8') as f2:
        json.dump(result, f2, ensure_ascii=False, indent=4)
    print('j')
```

```python
count_trends('/yy/trends/CS_integrate_trends.json','/yy/trends/CS_count_trends.json')
count_trends('/yy/trends/LIS_integrate_trends.json','/yy/trends/LIS_count_trends.json')
count_trends('/yy/trends/MI_integrate_trends.json','/yy/trends/MI_count_trends.json')
count_trends('/yy/trends/MS_integrate_trends.json','/yy/trends/MS_count_trends.json')
count_trends('/yy/trends/physics_integrate_trends.json','/yy/trends/physics_count_trends.json')
```
