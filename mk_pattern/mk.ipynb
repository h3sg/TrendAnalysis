```python
import pymannkendall as mk
import json
```

```python
def return_predict_trend(file, alpha_value, keyword_index=None):
    result_list = []
    with open(file, "r", encoding="utf-8") as f:
        lines = f.readlines()
    for line in lines:
        parts = line.strip().split(" ")
        kw = parts[0]
        counts = list(map(float,parts[1:]))
        result_list.append((kw, counts))
    trends = {}
    for keyword, count in result_list:
        result = mk.original_test(count, alpha=alpha_value)
        if result.trend == "increasing":
            trend = "上升趋势"
        elif result.trend == "decreasing":
            trend = "下降趋势"
        else:
            trend = "没有趋势"
        is_significant = result.p <= alpha_value
        '''trends[idx] = {
            "趋势": trend,
            "显著性": is_significant,
            "p值": result .p,
            "z值": result.z
        }'''
        trends[keyword] = trend
    return trends
    '''if keyword_index is None:
        print(trends)
    if keyword_index is not None and keyword_index in trends:
        print(trends[keyword_index])
    else:
        print("输入的关键词索引不存在")'''
```

```python
def save_trends(trends,file):
    with open(file,'w',encoding='utf-8') as f:
        json.dump(trends,f,indent=4,ensure_ascii=False)
    print(f'{file}趋势已保存！')
```

```python
file_path1 = '/yy/MK_pattern/word_count.txt'
word_count_trend = return_predict_trend(file_path1,0.05)
save_trends(word_count_trend,'/yy/MK_pattern/word_count_trend.json')
```
