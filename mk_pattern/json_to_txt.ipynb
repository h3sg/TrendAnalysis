```python
import json
```

```python
def json_to_txt1(readfile_path,writefile_path):
    with open(readfile_path,"r",encoding="utf-8") as r_file:
        data = json.load(r_file)
    with open(writefile_path,"w",encoding="utf-8") as w_file:
        for idx,kw in enumerate(data.keys()):
            w_file.write(kw + " ")
            for i in range(0,13):
                w_file.write(str(data[kw][i]) + " ")
            w_file.write(str(data[kw][-1]) + "\n")
    print("json剌쒾즲蚌섟쟠txt")
```

```python
readfile_path = "/yy/MK_pattern/keyword_sum.json"
writefile_path = "/yy/MK_pattern/word_count.txt"
json_to_txt1(readfile_path,writefile_path)
```

```python
readfile_path = "/yy/MK_pattern/keyword_together_degree.json"
writefile_path = "/yy/MK_pattern/keyword_together_degree.txt"
json_to_txt1(readfile_path,writefile_path)
```
