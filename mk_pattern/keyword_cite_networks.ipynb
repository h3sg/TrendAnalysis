```python
def create_doi_to_keywords_mapping(papers):
    doi_to_keywords = {}
    for paper in papers:
        doi = paper.get("DOI", "")
        keywords = paper.get("KW", [])
        if doi:
            doi_to_keywords[doi] = keywords
    return doi_to_keywords
```

```python
def create_keyword_network(papers, doi_to_keywords):
    G = nx.DiGraph()  # 创建有向图

    for paper in papers:
        doi = paper.get("DOI", "")  # 当前论文的 DOI
        keywords = paper.get("KW", [])  # 当前论文的关键词
        citations = paper.get("CI", [])  # 参考文献的 DOI 列表

        # 遍历每篇参考文献
        for citation_doi in citations:
            # 如果引用的参考文献 DOI 存在于文件中，则继续处理
            if citation_doi in doi_to_keywords:
                cited_keywords = doi_to_keywords[citation_doi]

                # 参考文献中的关键词指向当前论文的关键词
                for cited_kw in cited_keywords:
                    for kw in keywords:
                        G.add_edge(kw, cited_kw)  # 添加有向边，当前论文关键词 -> 参考文献关键词

    return G
```

```python
def compute_in_out_degrees(G):
    degree_list = []

    for node in G.nodes():
        in_degree = G.in_degree(node)  # 计算入度
        out_degree = G.out_degree(node)  # 计算出度
        degree_list.append({
            node: {
                "入度": in_degree,
                "出度": out_degree
            }
        })

    return degree_list
```

```python
def save_in_out_degrees_to_json(degree_list, output_path):
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(degree_list, f, ensure_ascii=False, indent=4)
    print(f"节点的入度和出度已保存到 {output_path}")
```

```python
doi_to_keyword_lists = [{} for _ in range(14)]
print(doi_to_keyword_lists)
```

```python
for i in range(14):
    doi_to_keyword_lists[i] = create_doi_to_keywords_mapping(lists[i])
```

```python
graphs = []
for j in range(14):
    graph = create_keyword_network(lists[j],doi_to_keyword_lists[j])
    graphs.append(graph)
print(graphs[0])
```

```python
in_out_degree_lists = []
for k in range(14):
    in_out_degree_lists.append(compute_in_out_degrees(graphs[k]))
print(in_out_degree_lists[0][0])
```

```python
keyword_cite_networks = {}
for idx,year_data in enumerate(in_out_degree_lists):
    for entry in year_data:
        for kw_node,in_out_deg in entry.items():
            if kw_node not in keyword_cite_networks:
                keyword_cite_networks[kw_node] = [[0,0] for _ in range(14)]
            keyword_cite_networks[kw_node][idx][0] = in_out_deg["入度"]
            keyword_cite_networks[kw_node][idx][1] = in_out_deg["出度"]
output_path = "/hy-tmp/pattern/keyword_cite_in_out_degree.json"
save_in_out_degrees_to_json(keyword_cite_networks,output_path)
```
