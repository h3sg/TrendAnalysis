```python
with open('data/kw_p_m.txt','r',encoding='utf-8') as f1:
    lines=f1.readlines()
word_list=[]
for word in lines:
    word_list.append(word.strip())
# word_list[:3]
len(word_list)
```

```python
# 获取2010-2017，2018-2022，2023三个时间段的关键词，并以txt格式保存
import json
from datetime import datetime
import pandas as pd

start_time=datetime.now()
with open('/yy/pattern/CS/wos_CS.json','r',encoding='utf-8') as f2:
    data=json.load(f2)
word_list1=[]
word_list2=[]
word_list3=[]
for word in word_list:
    for item in data:
        if (word in item['KW']) and (int(item['YE'])>=2010) and (int(item['YE'])<=2017):
            word_list1.append(word)
        elif (word in item['KW']) and (int(item['YE'])>=2018) and (int(item['YE'])<=2022):
            word_list2.append(word)
        elif (word in item['KW']) and (int(item['YE'])==2023):
            word_list3.append(word)
# 去重
word_list_2010_2017=list(set(word_list1))
word_list_2018_2022=list(set(word_list2))
word_list_2023=list(set(word_list3))
# 保存文件
with open('data/word_2010_2017.txt','w',encoding='utf-8') as wf1:
    for word in word_list_2010_2017:
        wf1.write(word+'\n')
print(f'文件已保存至data/word_2010_2017.txt，长度为{len(word_list_2010_2017)}')

with open('data/word_2018_2022.txt','w',encoding='utf-8') as wf2:
    for word in word_list_2018_2022:
        wf2.write(word+'\n')
print(f'文件已保存至data/word_2018_2022.txt，长度为{len(word_list_2018_2022)}')

with open('data/word_2023.txt','w',encoding='utf-8') as wf3:
    for word in word_list_2023:
        wf3.write(word+'\n')
print(f'文件已保存至data/word_2023.txt，长度为{len(word_list_2023)}')
end_time=datetime.now()
print(f'耗时{end_time-start_time}')
```
