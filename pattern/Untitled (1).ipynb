```python
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
with open('data/TSP_co_m.txt','r',encoding='utf-8') as f:
    lines=f.readlines()
co_m=[]
for word in lines:
    co_m.append(word.strip())
# co_m[:3]
co_series={}
with open('data/problem_method.json','r',encoding='utf-8') as f1:
    data=json.load(f1)
for kw in co_m:
    co_series[kw]=data[kw]
co_series
```

```python
import numpy as np
# 年份
example={}
for word in list(co_series.keys())[:25]:
    example[word]=co_series[word]
example['NeuralNetworks']=co_series['NeuralNetworks']
example['FeatureSelection']=co_series['FeatureSelection']
example['Optimization']=co_series['Optimization']
years = list(range(2010, 2024))

# 转换为 DataFrame
df = pd.DataFrame(example, index=years)

# 对数归一化（防止数据悬殊）
df_log = np.log1p(df)

# 画布尺寸（优化可读性）
plt.figure(figsize=(18, 12), dpi=300)

# 画热力图
sns.heatmap(df_log.T, cmap="RdBu_r", linewidths=2, annot=True, fmt=".1f",
            annot_kws={"size": 10}, cbar=True, cbar_kws={'shrink': 0.6, 'label': 'Log Scale Frequency'})

# 轴标签优化
plt.xticks(rotation=45, fontsize=14)
plt.yticks(fontsize=14)

# 设置标题 & 轴标签
plt.title("2010-2023 关键词词频时间分布热力图", fontsize=18)
plt.xlabel("Year", fontsize=16)
plt.ylabel("Keywords", fontsize=16)

# 显示图像
plt.show()
```
